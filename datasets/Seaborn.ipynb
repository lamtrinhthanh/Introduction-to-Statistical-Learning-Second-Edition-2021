```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

## Visualize Linear Relationships

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.set(color_codes=True)
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='size', y='tip', data=tips)
sns.lmplot(x='size', y='tip', data=tips, x_estimator=np.median, x_ci=100)
```

```python
sns.lmplot(x='size', y='tip', data=tips, x_estimator=np.mean, x_ci=100)
plt.show()
```

```python
anscome = sns.load_dataset("anscombe")
```

```python
sns.lmplot(x='x', y='y', data=anscome.query("dataset=='II'"), order=2, ci=None, scatter_kws={'s':89})
plt.show()
```

```python
sns.lmplot(x='x', y='y', data=anscome.query("dataset=='III'"), ci=None, scatter_kws={'s':100}, robust=False)
sns.lmplot(x='x', y='y', data=anscome.query("dataset=='III'"), ci=None, scatter_kws={'s':100}, robust=True)
plt.show()
```

```python
#When the y variable is binary, simple linear regression also â€œworksâ€ but provides implausible predictions:
tips["big_tips"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x='total_bill', y='big_tips', data=tips, ci=None, scatter_kws={'s':80})
#logistic regression
sns.lmplot(x='total_bill', y='big_tips', data=tips, ci=None, logistic=True, scatter_kws={'s': 80})
#nonparametric regression
sns.lmplot(x='total_bill', y='big_tips', data=tips, ci=None, lowess=True, scatter_kws={'s':80})
plt.show()
```

```python
#conditioning on the third variable
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', ci=None, scatter_kws={'s':80})
#change color set and marker shape
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', ci=None, scatter_kws={'s':80}, markers=['o', 'x'], palette='muted')
plt.show()
```

```python
#draw multiple â€œfacetsâ€ which each level of the variable appearing in the rows or columns of the grid
#by column
sns.lmplot(x="total_bill", y="tip", data=tips, hue='smoker', col='sex', scatter_kws={'s':100})
#by row
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', row='time', col='sex', scatter_kws={'s':100})
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col='smoker', col_wrap=2, height=5, aspect=1)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col='smoker', col_wrap=5)
```

```python
#pairplot
sns.pairplot(data=tips, hue='smoker', x_vars=['total_bill','size'], y_vars=['tip','sex'], aspect=1, height=5, plot_kws={'s':80})
plt.show()
```
