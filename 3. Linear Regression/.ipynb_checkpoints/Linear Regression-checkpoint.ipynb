### 3.1.1 Estimating the Coefficients

```python
#use sklearn to fit a least squares line to model and matplotlib to visualize the data and fitting result
linreg = LinearRegression()
linreg.fit(advertising.TV.to_numpy().reshape(-1, 1), advertising.Sales.to_numpy().reshape(-1, 1))
predictions = linreg.predict(advertising.TV.to_numpy().reshape(-1, 1))

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.plot(advertising.TV, advertising.Sales, 'ro')
ax.plot(advertising.TV, predictions, 'b-', linewidth=2)
ax.plot((advertising.TV, advertising.TV), (advertising.Sales, predictions.flatten()), c='brown', linewidth=0.5)
ax.set_xlabel('TV', fontsize=20)
ax.set_ylabel('Sales', fontsize=20)
ax.set_title('Figure 1', fontsize=20);
```

```python
print("The coefficients:\nB0 = {}\nB1 = {}".format(linreg.intercept_.item(), linreg.coef_.item()))
```

```python
#use seaborn
ax = sns.regplot(x="TV", y="Sales", data=advertising, scatter_kws={"color":"red"}, line_kws={"color":"blue"}, ci=None)
ax.figure.set_size_inches((10, 8))
ax.yaxis.label.set_size(20)
plt.setp(ax.xaxis.get_label(), size=20)
ax.set_title("Figure 1", fontsize=20);
```

```python
#make a contour plot of RSS wrt B0, B1
B0 = np.linspace(linreg.intercept_ - 2, linreg.intercept_ + 2, 100)
B1 = np.linspace(linreg.coef_ - 0.02, linreg.coef_ + 0.02, 100)
xx, yy = np.meshgrid(B0, B1)
RSS = np.zeros(xx.shape)
for (i, j), _ in np.ndenumerate(RSS):
    RSS[i, j] = np.sum(np.square(yy[i, j] * advertising.TV + xx[i, j] - advertising.Sales)) / 1000

fig = plt.figure(figsize=(15, 8))
fig.subplots_adjust()
ax1 = fig.add_subplot(1, 2, 1)
cs = ax1.contour(xx, yy, RSS, levels=[2.15, 2.2, 2.3, 2.5, 3], colors='b', alpha=0.4)
ax1.scatter(linreg.intercept_, linreg.coef_, c='red', s=40)
ax.clabel(cs, fmt="%1.1f", inline=True)
ax2 = fig.add_subplot(122, projection="3d")
ax1.set_xlabel(r'$\beta_0$', fontsize=15)
ax1.set_ylabel(r'$\beta_1$', fontsize=15, rotation=0)
cs = ax2.contour(xx, yy, RSS, levels=[2.15, 2.2, 2.3, 2.5, 3], offset=RSS.min(),  alpha=0.5)
ax2.plot_surface(xx, yy, RSS, alpha=0.4)
ax2.scatter3D(linreg.intercept_, linreg.coef_, RSS.min(), color='r')
fig.suptitle('RRS - Regression Coefficients', fontsize=20)
ax2.clabel(cs, fmt="%1.1f")
ax2.set_xlabel(r'$\beta_0$', fontsize=15)
ax2.set_ylabel(r'$\beta_1$', fontsize=15)
ax2.set_zlabel('RSS', fontsize=15, rotation=0);
```

### 3.1.2 Assessing the Accuracy of the Coefficient Estimates

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

est = smf.ols(formula="Sales ~ TV", data=advertising).fit();
est.summary().tables[1]
```

We see that the the values of $\beta_0$ and $\beta_1$ are large relative to their standard error, leading to large t-statistics and almost 0 for p-values

```python
est.summary()
```

Residual Standard Error ${RSE}$ and ${R^2}$

```python
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
X, y = advertising.TV.to_numpy().reshape(-1, 1), advertising.Sales
linreg = LinearRegression()
linreg.fit(X, y)
y_pred = linreg.predict(X)
display(
    pd.DataFrame([["Residual Standard Error",
                   mean_squared_error(y, y_pred)],
                  ["${R^2}$", r2_score(y, y_pred)]],
                 columns=["Metric", "Value"]).set_index('Metric'))
```

```python
X_limage = credit[["Limit", "Age"]].values
X_limrat = credit[["Limit", "Rating"]].values
Y = credit.Balance.to_numpy().reshape(-1, 1)
```

```python
from sklearn.preprocessing import scale
X_limage = scale(X_limage, with_std=False)
X_limrat = scale(X_limrat, with_std=False)
```

```python
linreg = LinearRegression()
linreg.fit(X_limage, Y)
B_limit1, B_age = linreg.coef_.squeeze().tolist()
intercept_1 = linreg.intercept_
```

```python
import warnings; warnings.simplefilter('ignore')
```

```python
#estimate coefficients in the least square fit including only two predictors limit and age and the response balance
linreg1 = LinearRegression()
linreg1.fit(X_limage, Y)
B_limit1, B_age = linreg1.coef_.squeeze().tolist()
intercept_1 = linreg1.intercept_
#estimate coefficients in the least square fit including only two predictors limit and rating and the response balance
linreg2 = LinearRegression()
linreg2.fit(X_limrat, Y)
B_limit2, B_rating = linreg2.coef_.squeeze().tolist()
intercept_2 = linreg2.intercept_
RSS = lambda B_1, B_2, X, Y, intercept_: np.sum(np.square((B_1 * X[:, 0] + B_2 * X[:, 1] + intercept_) - Y.flatten())) / 1000000
#Contour plot of RSS for the regression of balance onto age and limit
X1 = np.linspace(B_limit1 - 0.03,  B_limit1 + 0.03, 100)
Y1 = np.linspace(B_age - 3, B_age + 3, 100)
xx1,  yy1 = np.meshgrid(X1, Y1)
Z1 = np.zeros(xx1.shape)
for (i, j), _ in np.ndenumerate(Z1):
    Z1[i, j] = ((Y.flatten() - (xx1[i, j] * X_limage[:, 0] + yy1[i, j] * X_limage[:, 1] + intercept_1))**2).sum() / 1000000
    
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
CS = ax[0].contour(xx1, yy1, Z1, colors="dodgerblue", levels=[21.25, 21.5, 21.8])
#find the indexes of the lowest cost
r, c = np.where(Z1 == np.min(Z1))
best_B_limit1 = xx1[r[0], c[0]]
best_B_age = yy1[r[0], c[0]]
ax[0].scatter(best_B_limit1, best_B_age, c="k")
ax[0].plot((best_B_limit1, best_B_limit1), (ax[0].get_ylim()[0], best_B_age), 'k--', linewidth=0.5)
ax[0].plot((best_B_limit1, ax[0].get_xlim()[0]), (best_B_age, best_B_age), 'k--', linewidth=0.5)
ax[0].set(xlabel=r"$\beta_{limit}$", ylabel=r"$\beta_{age}$")
ax[0].xaxis.label.set_size(15)
ax[0].yaxis.label.set_size(15)
ax[0].set_yticklabels(ax[0].get_yticks(), fontsize=15)
ax[0].set_xticklabels(np.round(ax[0].get_xticks(), 2), fontsize=15)
ax[0].clabel(CS, inline=True, fmt="%1.1f", fontsize=15);
fig.suptitle("Figure 3.15", fontsize=25, y=1)
#Contour plot of RSS for the regression of balance onto limit and rating
X2 = np.linspace(B_limit2 - 0.2, B_limit2 + 0.2, 100)
Y2 = np.linspace(B_rating - 3, B_rating + 3, 100)
xx2, yy2 = np.meshgrid(X2, Y2)
Z2 = np.zeros(xx2.shape)
for (i, j), _ in np.ndenumerate(Z2):
    Z2[i, j] = ((Y.flatten() - (xx2[i, j] * X_limrat[:, 0] + yy2[i, j] * X_limrat[:, 1] + intercept_2))**2).sum() / 1000000
r, c = np.where(Z2 == np.min(Z2))
best_B_limit2 = xx2[r[0], c[0]]
best_B_rating = yy2[r[0], c[0]]
ax[1].scatter(best_B_limit2, best_B_rating, c='k')
CS = ax[1].contour(xx2, yy2, Z2, levels=[21.5, 21.8], colors="blue")
ax[1].plot((best_B_limit2, best_B_limit2), (ax[1].get_ylim()[0], best_B_rating), 'k--', linewidth=0.5)
ax[1].plot((best_B_limit2, ax[1].get_xlim()[0]), (best_B_rating, best_B_rating), 'k--', linewidth=0.5)
ax[1].clabel(CS, fmt="%1.1f")
ax[1].set(xlabel=r"$\beta_{limit}$", ylabel=r"$\beta_{rating}$")
ax[1].xaxis.label.set_size(15)
ax[1].yaxis.label.set_size(15)
ax[1].set_yticklabels(ax[0].get_yticks(), fontsize=15)
ax[1].set_xticklabels(np.round(ax[0].get_xticks(), 2), fontsize=15)
plt.subplots_adjust(wspace=0.5);
```

```python
est = smf.ols("medv ~ {}".format(string_cols), data=boston).fit()
est.summary()
```

```python
#interaction term added 
est = smf.ols("medv ~ lstat * age", data=boston).fit()
est.summary()
```

```python
#add polynomial term
est = smf.ols("medv ~ lstat + np.power(lstat, 2)", data=boston).fit()
est.summary()
```

```python
from statsmodels.stats.anova import anova_lm
results1 = smf.ols("medv ~ lstat", data=boston).fit()
results2 = smf.ols("medv ~ lstat + np.square(lstat)", data=boston).fit()

anova_lm(results1, results2)
```

As there are two models, 1 is simple least square fit and 2 is the 2-nd degree polynomial fit for regression of medv onto lstat. It is next to test if the complex model is superior to the other or not. This can be done via anova test. The $H_{0}$ claims that no difference in the two models while $H_{1}$ is that the polinomial fit is better. As can be seen from the result, F=136 associated with p-value = 0 allows to reject the null hypothesis and provides very clear evidence that the second model is much better.
